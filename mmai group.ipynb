```python
cols_of_interest = [
  'F_AGECAT','F_GENDER','F_EDUCCAT2','F_INC_TIER2','F_PARTYLN_FINAL','INDUSTRYCOMBO_W119',
  'AI_HEARD_W119','USEAI_W119','TWITTER1_MOD_W119','AINEWS1_W119','AINEWS2_W119',
  'AIKNOW_INDEX_W119','DESRISK_COMF_W119','AIHCCOMF_W119','AIHCTRT1_W119','AIHCTRT2_W119',
  'AIWRKH1_W119','WEIGHT_W119'
]
df_core = df[cols_of_interest]
```

```python
print(df_core['AINEWS2_W119'].value_counts(dropna=False))
```

```python
for c in df_core.select_dtypes(include='object'):
    df_core[c] = df_core[c].replace(r'^\s*$', np.nan, regex=True)

df_core = df_core.replace(99, np.nan)

for c in ['F_AGECAT','F_GENDER','F_EDUCCAT2']:
    mode_val = df_core[c].mode(dropna=True).iloc[0]
    df_core[c] = df_core[c].fillna(mode_val)

df_core['F_INC_TIER2'] = pd.to_numeric(df_core['F_INC_TIER2'], errors='coerce')
median_inc = df_core['F_INC_TIER2'].median()
df_core['F_INC_TIER2'] = df_core['F_INC_TIER2'].fillna(median_inc)

df_core['F_PARTYLN_FINAL']    = df_core['F_PARTYLN_FINAL'].fillna('Unknown')
df_core['INDUSTRYCOMBO_W119'] = df_core['INDUSTRYCOMBO_W119'].fillna('Unknown')

for c in ['AI_HEARD_W119','USEAI_W119','TWITTER1_MOD_W119','AINEWS1_W119','AINEWS2_W119']:
    df_core[c] = df_core[c].fillna(0)

scale_cols = ['DESRISK_COMF_W119','AIHCCOMF_W119',
              'AIHCTRT1_W119','AIHCTRT2_W119','AIWRKH1_W119']
for c in scale_cols:
    df_core[c] = pd.to_numeric(df_core[c], errors='coerce')
    med = df_core[c].median()
    df_core[c] = df_core[c].fillna(med)


missing = df_core.isna().sum().to_frame('missing_count')
missing['missing_ratio'] = missing['missing_count'] / len(df_core)
print(missing)
```
